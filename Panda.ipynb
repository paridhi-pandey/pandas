## Load a sample CSV File

```python
import pandas as pd
#Loading the CSV file from your Google Drive
file_path = '/content/drive/MyDrive/salaries.csv'
df = pd.read_csv(file_path)

df.head()
```

## Cleaning missing values

```python
df.dropna()
```

## Taking out the Summary Statistics

```python
df.describe(include='all')
```

```python
df[df["salary"]>90000]
```

```python
df.groupby('work_year')['salary'].mean()
```
